```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/comple_n1.csv")
```

```python
df
```

```python
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(strategy="mean")
df[["Cutoff_2019","Cutoff_2020","Cutoff_2021","Cutoff_2022","Cutoff_2023","Cutoff_2024"]]=imputer.fit_transform(df[["Cutoff_2019","Cutoff_2020","Cutoff_2021","Cutoff_2022","Cutoff_2023","Cutoff_2024"]])
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

cat=ColumnTransformer([
    ("cat",OneHotEncoder(),["COLLEGE","BRANCH","CATEGORIES"])
])
df_cat=cat.fit_transform(df)
df_cat.shape
```

```python
x=df.drop(columns=["COLLEGE","BRANCH","CATEGORIES","Cutoff_2024"])
y=df["Cutoff_2024"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_predict=ridge.predict(x_test)

r2_score(y_predict,y_test)
```

```python
x
```

```python
df1=pd.DataFrame(df_cat.toarray())
```

```python
df_combined = pd.concat([x,df1], axis=1)
df_combined
```

```python
df_combined.columns = df_combined.columns.astype(str)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_combined,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(x_train,y_train)
```

```python
y_predict=ridge.predict(x_test)
r2_score(y_test,y_predict)
```

```python
from sklearn.linear_model import Lasso,LinearRegression
lasso=Lasso()
lasso.fit(x_train,y_train)
```

```python
y_predict=lasso.predict(x_test)
r2_score(y_test,y_predict)
```

```python
linear=LinearRegression()
linear.fit(x_train,y_train)
y_predict=linear.predict(x_test)
r2_score(y_test,y_predict)
```

```python
from sklearn.tree import DecisionTreeRegressor
trees=DecisionTreeRegressor()
trees.fit(x_train,y_train)
y_predict=trees.predict(x_test)
r2_score(y_test,y_predict)
```

```python
from sklearn.ensemble import RandomForestRegressor
trees=RandomForestRegressor()
trees.fit(x_train,y_train)
y_predict=trees.predict(x_test)
r2_score(y_test,y_predict)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Define the parameter grid
param_dist = {
    'n_estimators': np.arange(50, 300, 50),
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Initialize model
rf = RandomForestClassifier()

# Randomized search
rf_random = RandomizedSearchCV(rf, param_dist, n_iter=50, cv=5, verbose=2, n_jobs=-1)
rf_random.fit(x_train, y_train)
```

```python
y_predict=rf_random.predict(x_test)
r2_score(y_test,y_predict)
```

```python
rf=RandomForestRegressor(n_estimators=150,min_samples_split=2,max_features='log2',max_depth=30,bootstrap=True)
rf.fit(x_train,y_train)
y_predict=rf.predict(x_test)
r2_score(y_test,y_predict)
```

```python
college = input("Enter the College name? ")
branch = input("Enter the Branch name? ")
category = input("Enter the Category? ") 
past_cutoffs_row = df.loc[
            (df['COLLEGE'] == college) & 
            (df['BRANCH'] == branch) & 
            (df['CATEGORIES'] == category)
        ]
past_cutoffs = past_cutoffs_row.iloc[0][['Cutoff_2019', 'Cutoff_2020', 'Cutoff_2021', 'Cutoff_2022', 'Cutoff_2023',]].tolist()
print(past_cutoffs)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
process = ColumnTransformer([
    ('convert', OneHotEncoder(handle_unknown='ignore'), ['COLLEGE', 'BRANCH', 'CATEGORIES'])
])

# Create the pipeline
model_pipeline = Pipeline([
    ("process", process),
    ("model", RandomForestRegressor())
])
imputer = SimpleImputer(strategy='mean')
df[['Cutoff_2019', 'Cutoff_2020', 'Cutoff_2021', 'Cutoff_2022', 'Cutoff_2023']] = imputer.fit_transform(
    df[['Cutoff_2019', 'Cutoff_2020', 'Cutoff_2021', 'Cutoff_2022', 'Cutoff_2023']]
)
X = df[['COLLEGE', 'BRANCH', 'CATEGORIES', 'Cutoff_2019', 'Cutoff_2020', 'Cutoff_2021', 'Cutoff_2022', 'Cutoff_2023']]
y = df['Cutoff_2024']

model_pipeline.fit(X,y)
```

```python
 input_data = pd.DataFrame({
            'COLLEGE': [college],
            'BRANCH': [branch],
            'CATEGORIES': [category],
            'Cutoff_2019': [past_cutoffs[0]],
            'Cutoff_2020': [past_cutoffs[1]],
            'Cutoff_2021': [past_cutoffs[2]],
            'Cutoff_2022': [past_cutoffs[3]],
            'Cutoff_2023': [past_cutoffs[4]],
        })
```
