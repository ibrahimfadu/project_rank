```python
college = input("Enter the College name? ")
branch = input("Enter the Branch name? ")
category = input("Enter the Category? ") 
past_cutoffs_row = df.loc[
            (df['COLLEGE'] == college) & 
            (df['BRANCH'] == branch) & 
            (df['CATEGORIES'] == category)
        ]
past_cutoffs = past_cutoffs_row.iloc[0][['Cutoff_2019', 'Cutoff_2020', 'Cutoff_2021', 'Cutoff_2022', 'Cutoff_2023',]].tolist()
print(past_cutoffs)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
process = ColumnTransformer([
    ('convert', OneHotEncoder(handle_unknown='ignore'), ['COLLEGE', 'BRANCH', 'CATEGORIES'])
])

# Create the pipeline
model_pipeline = Pipeline([
    ("process", process),
    ("model", RandomForestRegressor())
])
imputer = SimpleImputer(strategy='mean')
df[['Cutoff_2019', 'Cutoff_2020', 'Cutoff_2021', 'Cutoff_2022', 'Cutoff_2023']] = imputer.fit_transform(
    df[['Cutoff_2019', 'Cutoff_2020', 'Cutoff_2021', 'Cutoff_2022', 'Cutoff_2023']]
)
X = df[['COLLEGE', 'BRANCH', 'CATEGORIES', 'Cutoff_2019', 'Cutoff_2020', 'Cutoff_2021', 'Cutoff_2022', 'Cutoff_2023']]
y = df['Cutoff_2024']

model_pipeline.fit(X,y)
```
