```python
import pandas as pd
import math
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('data/data.csv')

def classify_tier(user_rank, cutoff):
    """
    Classify the recommendation into tiers based on the difference between cutoff and user's rank.
    
    - "Safe": if the user's rank is significantly better than the cutoff
    - "Match": if the user's rank is moderately better than the cutoff
    - "Reach": if the user's rank is just within the cutoff
    (Thresholds are illustrative; adjust based on your domain knowledge.)
    """
    difference = cutoff - user_rank
    if difference >= 200:
        return "Safe"
    elif 50 <= difference < 200:
        return "Match"
    elif 0 <= difference < 50:
        return "Reach"
    else:
        return "Not Eligible"  # Should not occur if we only recommend when user_rank <= cutoff

def calculate_probability(user_rank, cutoff, scale=50):
    """
    Calculate a probability percentage using a logistic function.
    
    The logistic function is defined as:
      probability = 1 / (1 + exp((user_rank - cutoff) / scale))
      
    For user_rank == cutoff, probability is 50%. As the rank improves (decreases),
    the probability approaches 100%.
    
    Parameters:
      user_rank (int): The DCET rank of the user (lower is better).
      cutoff (int or float): The cutoff value for the college.
      scale (float): Controls the steepness of the logistic function.
      
    Returns:
      float: The probability of admission as a percentage.
    """
    # If cutoff is 0, it indicates no seats available
    if cutoff == 0:
        return 0
    probability = 1 / (1 + math.exp((user_rank - cutoff) / scale))
    return probability * 100

def recommend_colleges(user_rank, user_category, user_branch=None, preferred_colleges=None, reference_year='Cutoff_2024'):
    """
    Recommend colleges based on user's DCET rank, category, and optionally branch and preferred colleges.
    
    For each recommendation, calculate:
      - The tier ("Safe", "Match", "Reach") based on the difference between the cutoff and user rank.
      - A chance percentage using a logistic function.
    
    The final recommendations are sorted in ascending order by the cutoff value.
    
    Parameters:
      user_rank (int): The DCET rank of the user (lower is better).
      user_category (str): The applicant category.
      user_branch (str, optional): The branch preference provided by the user.
      preferred_colleges (list, optional): A list of college names to filter recommendations.
      reference_year (str): Which year's cutoff data to use (default is 'Cutoff_2024').
      
    Returns:
      A list of recommendations where each recommendation is a dictionary containing:
      - College name
      - Branch
      - Cutoff value used for comparison
      - Tier classification (Safe, Match, Reach)
      - Chance percentage of admission
    """
    recommendations = []
    
    # Iterate over each record in the dataset
    for _, row in df.iterrows():
        # Get the cutoff value for the reference year
        cutoff = row[reference_year]
        
        # Skip if cutoff is 0 (indicates no seats available for that branch/category)
        if cutoff == 0:
            continue
        
        # Filter by the applicant's category
        if row['CATEGORIES'] != user_category:
            continue
        
        # Filter by the branch if the user provided one
        if user_branch and row['BRANCH'] != user_branch:
            continue
        
        # If preferred_colleges is provided, filter out rows not in the list
        if preferred_colleges and row['COLLEGE'] not in preferred_colleges:
            continue
        
        # Recommend if the user's rank meets or beats the cutoff (remember: lower rank is better)
        if user_rank <= cutoff:
            tier = classify_tier(user_rank, cutoff)
            chance = calculate_probability(user_rank, cutoff)
            recommendations.append({
                'College': row['COLLEGE'],
                'Branch': row['BRANCH'],
                'Cutoff': cutoff,
                'Tier': tier,
                'Chance (%)': round(chance, 2)
            })
    
    # Remove duplicate recommendations by college name (if needed)
    unique_recommendations = {rec['College']: rec for rec in recommendations}.values()
    
    # Sort recommendations by cutoff in ascending order
    sorted_recommendations = sorted(unique_recommendations, key=lambda rec: rec['Cutoff'])
    
    return sorted_recommendations

def plot_recommendations(recommendations):
    """
    Plot the recommended colleges using a bar chart for cutoff values and a line chart for admission chance.
    """
    if not recommendations:
        print("No recommendations to display.")
        return
    
    # Extract data from recommendations
    colleges = [rec['College'] for rec in recommendations]
    cutoff_values = [rec['Cutoff'] for rec in recommendations]
    chance_values = [rec['Chance (%)'] for rec in recommendations]
    
    # Set up the plot
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    # Plot cutoff values as bars
    bar_width = 0.5
    bars = ax1.bar(colleges, cutoff_values, bar_width, label="Cutoff", color='skyblue', alpha=0.8)
    ax1.set_ylabel('Cutoff Score', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    plt.xticks(rotation=45, ha="right")
    
    # Create a twin axis for the chance percentages
    ax2 = ax1.twinx()
    ax2.plot(colleges, chance_values, color='red', marker='o', label="Chance (%)")
    ax2.set_ylabel('Chance (%)', color='red')
    ax2.tick_params(axis='y', labelcolor='red')
    
    # Title and legend
    plt.title('College Recommendations: Cutoff vs Chance of Admission')
    fig.tight_layout()
    plt.legend(loc='upper left')
    plt.show()

# --------------------------
# Example usage:
# --------------------------

# Sample user inputs
user_rank = 13         # DCET rank provided by the user (lower is better)
user_category = 'GM'    # Applicant category provided by the user
user_branch = 'CS Computers'  # Branch preference provided by the user
preferred_colleges = None  # Optionally, a list of preferred colleges

# Get recommendations using the 2024 cutoff as reference
results = recommend_colleges(user_rank, user_category, user_branch, preferred_colleges, reference_year='Cutoff_2024')

print("Recommended Colleges based on your rank, category, and branch (sorted by cutoff ascending):\n")
if results:
    for rec in results:
        print(f"College: {rec['College']} | Branch: {rec['Branch']} | Cutoff: {rec['Cutoff']} | Tier: {rec['Tier']} | Chance: {rec['Chance (%)']}%")
    # Plot the recommendations
    plot_recommendations(results)
else:
    print("No colleges match your criteria based on the historical cutoff data.")
```

```python
import pandas as pd
import math
import numpy as np

# Load data from CSV file
df = pd.read_csv('data/data.csv')

def predict_cutoff_2025(row):
    """
    Predict the cutoff for 2025 for a given college record using its historical cutoff data.
    It fits a linear trend on the years 2019 to 2024 and extrapolates for 2025.
    
    Parameters:
      row (pd.Series): A row from the dataframe containing cutoff values.
      
    Returns:
      float: The predicted cutoff for 2025.
    """
    # Define the years and corresponding cutoff values
    years = np.array([2019, 2020, 2021, 2022, 2023, 2024])
    cutoffs = np.array([
        row['Cutoff_2019'], row['Cutoff_2020'], row['Cutoff_2021'], 
        row['Cutoff_2022'], row['Cutoff_2023'], row['Cutoff_2024']
    ])
    
    # If all cutoff values are zero (i.e., no seats available), return 0
    if np.all(cutoffs == 0):
        return 0
    
    # Fit a linear regression (1st degree polynomial) to the data points
    slope, intercept = np.polyfit(years, cutoffs, 1)
    # Predict the cutoff for the year 2025
    predicted = slope * 2025 + intercept
    return predicted

def classify_tier(user_rank, cutoff):
    """
    Classify the recommendation into tiers based on the difference between cutoff and user's rank.
      - "Safe": if the user's rank is significantly better than the cutoff
      - "Match": if the user's rank is moderately better than the cutoff
      - "Reach": if the user's rank is just within the cutoff
    """
    difference = cutoff - user_rank
    if difference >= 200:
        return "Safe"
    elif 50 <= difference < 200:
        return "Match"
    elif 0 <= difference < 50:
        return "Reach"
    else:
        return "Not Eligible"

def calculate_probability(user_rank, cutoff, scale=50):
    """
    Calculate the probability of admission using a logistic function.
    
    For user_rank equal to the cutoff, the probability is 50%.
    As the rank improves (decreases), the probability approaches 100%.
    """
    if cutoff == 0:
        return 0
    probability = 1 / (1 + math.exp((user_rank - cutoff) / scale))
    return probability * 100

def recommend_colleges(user_rank, user_category, user_branch=None, preferred_colleges=None, 
                       reference_year='Cutoff_2024', use_2025_prediction=False):
    """
    Recommend colleges based on the user's DCET rank, category, and optionally branch and preferred colleges.
    
    If use_2025_prediction is True, the function uses the predicted cutoff for 2025 (based on historical data)
    instead of the actual cutoff from the reference year.
    
    The recommendations are sorted in ascending order by cutoff.
    
    Returns:
      A list of recommendations, where each recommendation is a dictionary containing:
        - College name
        - Branch
        - Cutoff (or predicted cutoff)
        - Tier classification (Safe, Match, Reach)
        - Chance percentage of admission
    """
    recommendations = []
    
    for _, row in df.iterrows():
        # Choose between actual cutoff or predicted cutoff for 2025
        if use_2025_prediction:
            cutoff = predict_cutoff_2025(row)
        else:
            cutoff = row[reference_year]
        
        # Skip if cutoff is 0 (i.e., no seats available)
        if cutoff == 0:
            continue
        
        # Filter by the applicant's category
        if row['CATEGORIES'] != user_category:
            continue
        
        # Filter by the branch if provided
        if user_branch and row['BRANCH'] != user_branch:
            continue
        
        # Filter by preferred colleges if provided
        if preferred_colleges and row['COLLEGE'] not in preferred_colleges:
            continue
        
        # Recommend if the user's rank is within the cutoff (remember: lower rank is better)
        if user_rank <= cutoff:
            tier = classify_tier(user_rank, cutoff)
            chance = calculate_probability(user_rank, cutoff)
            recommendations.append({
                'College': row['COLLEGE'],
                'Branch': row['BRANCH'],
                'Cutoff': cutoff,
                'Tier': tier,
                'Chance (%)': round(chance, 2)
            })
    
    # Remove duplicate recommendations by college name (if needed)
    unique_recommendations = {rec['College']: rec for rec in recommendations}.values()
    
    # Sort recommendations by cutoff in ascending order
    sorted_recommendations = sorted(unique_recommendations, key=lambda rec: rec['Cutoff'])
    
    return sorted_recommendations

# --------------------------
# Example usage:
# --------------------------

# Sample user inputs
user_rank = 0                            # DCET rank provided by the user (lower is better)
user_category = 'GM'                        # Applicant category provided by the user
user_branch = 'CS Computers'   # Branch preference provided by the user
preferred_colleges = None                    # Optionally, a list of preferred colleges

# Get recommendations using predicted cutoff values for 2025
results = recommend_colleges(user_rank, user_category, user_branch, preferred_colleges,
                             reference_year='Cutoff_2024', use_2025_prediction=True)

print("Recommended Colleges based on your rank, category, and branch (using predicted 2025 cutoffs):\n")
if results:
    for rec in results:
        print(f"College: {rec['College']} | Branch: {rec['Branch']} | Predicted Cutoff for 2025: {rec['Cutoff']} | Tier: {rec['Tier']} | Chance: {rec['Chance (%)']}%")
else:
    print("No colleges match your criteria based on the historical cutoff data.")
```
